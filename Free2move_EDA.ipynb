**Cleaning data : duplicated values**


```python
df.duplicated().sum()
```

```python
df_cleaned = df.drop_duplicates()
df_cleaned
```

**Cleaning data : Casting types**

```python
df_cleaned.dtypes
```

```python
df_cleaned['CHARGED'] = df_cleaned['CHARGED'].astype('boolean')
df_cleaned['STARTED_TIME'] = pd.to_datetime(df_cleaned['STARTED_TIME'])
df_cleaned['FINISHED_TIME'] = pd.to_datetime(df_cleaned['FINISHED_TIME'])
df_cleaned.dtypes
```

```python
df_cleaned.head()
```

**Adding duration data for future analysis**

```python
df_cleaned['DURATION'] = df_cleaned['FINISHED_TIME'] - df_cleaned['STARTED_TIME']
```

```python
df_cleaned['DURATION']
```

```python
df_cleaned['DURATION_MIN'] = df_cleaned['DURATION'].dt.total_seconds() / 60
```

**Cleaning Data : handling null values**

```python
df_cleaned.isna().sum()
```

```python
df_cleaned[df_cleaned['CHARGELEVELSTART'].isna()]
```
